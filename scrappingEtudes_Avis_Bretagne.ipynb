Téléchargement des études

```python
dataEtude = pd.read_csv(cheminInfos+"infos_etude.csv", delimiter=";") # permet de relancer la cellule en cas d'interruption sans relancer les cellules précédentes.
idDone = list(dataEtude['id'])


cpt = 0 #compteur pour limité le nombre de fichier à récupérer
maxi = 500
for index, line in liensEtude.iterrows():
    # Vérification limite
    if(cpt > maxi):
        break
    cpt += 1
    
    
    
    #récupération de l'ID du fichier, son extension et son departement
    idDoc = line['id']
    if(idDoc == ''):
        break
    extension_EI = re.search('\.((?:\w{1,6}))\?a=true\s*$', line['url_etude'])[1]
    tmp = re.search('(22|29|35|56)(?:.|\s){0,10}$', line['titre_projet'])
    departement = "" if tmp==None else tmp[1]
    name_file_EI = re.search('\/?([^\/]+)\?a=true\s*$', line['url_etude'])[1]
    
    if(idDoc not in idDone):#Si le document n'a pas déjà été traité
        print("ITERATION "+str(cpt))
    
        #################################
        #récupération de l'etude d'impact
        #################################
        
        now = datetime.datetime.now()
        print("   Telechargement etude d'impact... depuis ", now.hour, "h", now.minute)
        driver.get(line['url_etude'])
        time.sleep(1)

            
        ########################
        #sauvegarde informations
        ########################
        print("   Sauvegarde des données (ne pas interrompre)... depuis ", now.hour, "h", now.minute)
        
            
        dataCSV = idDoc+";"+name_file_EI+";"+departement+"\n"
        file = open(cheminInfos+'infos_etude.csv','a')
        file.write(dataCSV)
        file.close()
        
        input()
print("END")
```
